・二次元ホークス過程の強度は以下で表せる。


$ν^{1}(t)=ν^{1}_0+n^{1}_1\sum_{i=1}^{N_1}(1/τ_1)e^{-(t-t^{1}_i)/τ_1}+n^{1}_2\sum_{j=1}^{N_2}(1/τ_2)e^{-(t-t^{2}_j)/τ_2}
ν^{2}(t)=ν^{2}_0+n^{2}_1\sum_{i=1}^{N_1}(1/τ_1)e^{-(t-t^{1}_i)/τ_1}+n^{2}_2\sum_{j=1}^{N_2}(1/τ_2)e^{-(t-t^{2}_j)/τ_2}$

アルゴリズム1.
時間伸縮理論を用いた、観察期間[0,T]における強度関数$\hat{ν}(t)$をもつ、自己励起NBD過程のシミュレーション手順:$\hat{ν}(t)=ν_0+n\sum_{i=1}^{\hat{N}(t)}m_ih(t-\hat{t}_i)$

Ensue:$\hat{ν}(t)=ν_0+\hat{z}_t$\
1.$i←1$
2.期待値１の指数分布Ex(1)に従う乱数Eを生成し、イベント発生時刻$t_i=\frac{E}{ν_0}$とする。\
3.イベントサイズの分布からイベントサイズ$m$を生成し、余剰強度$\hat{z}←\frac{nm}{τ}$とする。強度$\hat{ν}(t)=ν_0+\hat{z}$とし、$i←i+1$とする。\
4.以下を繰り返す。\
    　4.1.期待値１の指数分布Ex(1)に従う乱数Eを生成する。\
    　4.2.非線形関数$ν_0t'+\hat{z}_{i-1}(1-e^{-t'/τ})-E=0$を解き、$t_i←t_{i-1}+t'$とする。\
    　4.3.イベントサイズ$m$を生成し、余剰強度$\hat{z}_i←\hat{z}_{i-1}e^{-t/τ}+\frac{nm}{τ}$とする。強度$\hat{ν}(t)=ν_0+\hat{z}_i$とし、$i←i+1$とする。\
5.強度{$\hat{ν}_i$}とイベント発生時刻{$t_i$}を出力する。
